Creating a 'codeCommune' column to be able to merge

```python
mortality['codeCommune'] = mortality['depdom'] + mortality['comdom1']
```

```python
mortality.drop(columns=['depdom', 'comdom1'])
```

```python
mortality['codeCommune'] = pd.to_numeric(mortality['codeCommune'])
```

```python
ammoniac_2003.head()
```

```python
ammoniac_2003.codeCommune[0] == 14256
```

```python
df = pd.merge(mortality, ammoniac_2003,on='codeCommune')
```
